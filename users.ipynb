```python
# Create a new column
df['full_name'] = df['first_name'] + ' ' + df['last_name']
df['email_domain']= df['email'].str.split('@').str[-1]
df['website_domain'] = df['web'].str.split('//').str[-1].str.split('/').str[0]
df[['full_name' , 'email_domain', 'website_domain']].head()
```

```python
# Selected specific columns
selected_columns = df[['full_name', 'email_domain','website_domain' ]]
print(selected_columns.head())
```

```python
# selected filetred data
ca_users = df[df['state'] == 'CA']
ca_users.head()
```

```python
# sorting
sorted_users = df.sort_values(by = 'last_name')
sorted_users[['first_name', 'last_name' ]].head()
```

```python
len(ca_users)
```

```python
# Grouping state and counting users
state_counts = df.groupby('state').size().reset_index(name = 'user_count')
print(state_counts)
```
