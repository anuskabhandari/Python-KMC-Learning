```python
import matplotlib.pyplot as plt
```

```python
#Line plot
days = ['Monday' , 'Tuesday', 'Wednesday' , 'Thursday', 'Friday' ]
temperatures = [ 22,55,45,17,30]
plt.plot(days, temperatures, marker = 'o' , label = 'Temperature') # point haru ko lai denote o le garxa
plt.title('Temperature over a week')
plt.xlabel('Days')
plt.ylabel('Temperature(C)')
plt.legend()
plt.show()
```

```python
#Bar chart
products = ['Mobile' , 'Laptop', 'Tablet']
sales = [150,  300, 100]
plt.bar(products , sales , color = ['blue', 'green', 'orange'])
plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show() # show le better show garxa
```
