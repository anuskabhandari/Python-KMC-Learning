```python
#DataFrame with two-dimensional data with rows and columns
data={
    'Name' : ["Alice", "JON" , "sham"],
    'Age' : [20,20,40],
    'City': [ 'NewYork', 'LosANGLES', 'MNR']
}
```

```python
# Filtering DATAFRAME
filtered_df=df[df['Age']>28]
print(filtered_df)
```

```python
# Multiple Conditions
filtered_df = df[(df['Age']>15) & (df['City']== 'NewYork')]
print(filtered_df)
```

```python
# Multiple Conditions
filtered_df = df[(df['Age']>15) | (df['City']== 'NewYork')]
print(filtered_df)
```

```python
# Crud Add , Update , Delete
# ADDing a new column
df['Salary'] = [5000, 6000, 7000]
print(df)
```

```python
#Updating existing column
df['Salary'] = df['Salary'] + 5000
print(df)
```

```python
#updating existing column with condition
df.loc[df['Age'] > 30 , 'Salary'] = df['Salary'] +1000
print(df)
```

```python
# DEleteing a column
df.drop('Salary' , axis = 1 , inplace = True) ## column wise delete hunxa
print(df)
```

```python
# DEleteing a row
df.drop(0 , axis = 0 , inplace = True) ## row wise delete hunxa
print(df)
```
